```python
import numpy as np
# 1.NumPy stands for Numerical Python 2.In Python we have lists that serve the purpose of arrays, but they are slow to process.
#NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.
#he array xobject in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.
import pandas as pd
#Pandas is a Python library used for working with data sets
#to import and manipulate  and read data sets we use the pandas.
import difflib
#get close matches eg: if we enter any wrong name it will recommeds the relatable words.
from sklearn.metrics.pairwise  import cosine_similarity
#cosine similarity - we use to sepatarte the movies based on the cosine simillarity.those movies having highest similarity we recommed to the user 
from sklearn.feature_extraction.text import TfidfVectorizer
#string data to the  numperical data we use feature_extraction.
```

### i am going to use only selected colums from data.

```python
selected_list=['genres','keywords','tagline','original_title','cast','director']
print(selected_list)
```

## replacing null values in selected list

```python
# if there are any missing values we have to replace with the null vlaues.
for i in selected_list:
    movies_data[i]=movies_data[i].fillna('')
```

## combing all the selected features

```python
combine_data=movies_data['original_title']+' '+movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combine_data)
```

### we are converting the text data into the numericals beacause our cosine similarity is simple to caluculate with the numerical data



```python
#TfidfVectorizer tf-term frequency and idf-iverse document frequency
#tf =no of times term (t)  appered in the document/no of total terms i the document is the term frequency.
#idf=log(N/n)
#N = number of documents n=number of documents i which our t appers.
#for rare word the idf is high and frequent word my idf value is low.
vectorizer=TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combine_data)
print(feature_vectors)
```

## cosine similarity

```python
#now we will strore all the similar scores.
#how cosine similarity works.
#the each movie check with all the movies in the data and it will produce the similirity score.
similarity=cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
#getting the movie name from the user
movie_name=input('enter your movie name champ: ')
```

```python
#creating a list with all the movies given in the list
#selectint the titl column int the data.
title_list=movies_data['title']
#converting into list of movies.
title_list=movies_data['title'].tolist()
print(title_list)
```

```python
#find the close match.
find_close_match=difflib.get_close_matches(movie_name,title_list)
print(find_close_match)
```

```python
#best close match
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_the_movie=movies_data[movies_data.title==close_match]['index'].values[0]
#for geeting the actual index we will use values[0].bcz the data is in the form list for getting exact index we are using it.
print(index_of_the_movie)
```

```python
#geting the similar movies 
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
#first value is index and second value is the 
```

```python
#len of the similarity score
len(similarity_score)
```

```python
#print the similar movies based on the index
print("the suggested movies are")
i=0
for movie in similarity_score:
    index=movie[0]
    title_from_index=movies_data[movies_data.index ==index]['title'].values[0]
    if(i<30):
        print(i,'.',title_from_index)
    else:
        break
    i+=1;
```

```python
#sorting the movies based on higher similarity score
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#movie recomendation system
movie_name = input(' Enter your favourite movie name : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
